#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

#### Scatter Plot: A scatter plot represents individual data points as markers on a 2D coordinate system. It is useful for visualizing the distribution and correlation between two continuous variables.

```python
plt.scatter(x,y)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("X vs Y")
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.Use the following data:


```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

plt.tight_layout()  # Improves the spacing between subplots
plt.show()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


#### Box plots are used to visualize and compare the distributions of different datasets or variables. They provide information about the central tendency, spread, skewness, and presence of outliers within each dataset. Box plots are particularly useful for identifying differences in the distribution shape, variability, and skewness between groups or categories.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for box plots
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a single list
data = [box1, box2]

# Create a box plot
plt.boxplot(data)

# Set the x-axis tick labels
plt.xticks([1, 2], ['Box 1', 'Box 2'])

# Set the y-axis label
plt.ylabel('Values')

# Set the title
plt.title('Box Plot')

plt.show()
```
